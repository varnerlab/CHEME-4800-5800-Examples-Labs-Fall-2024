# Example: Grocery Fruit Problem using Cobb-Douglas Utility 
This example will familiarize students with solving a non-linear maximum utility problem in which we numerically estimate an optimal solution using methods from [the MadNPL library](https://github.com/MadNLP/MadNLP.jl). In this example, we are going to `Buy` versus `Build,` i.e., we have no idea what is going on inside [the MadNPL library](https://github.com/MadNLP/MadNLP.jl), we trust (and verify) the answer that it is giving us.

### Background
Consider a utility maximization problem with the Cobb-Douglas utility function, $U(x_{1},x_{2}) = x_{1}^{\alpha}x_{2}^{1-\alpha}$, subject to a budget constraint. 
If we were to do this with calculus, then [the Lagrangian](https://en.wikipedia.org/wiki/Lagrange_multiplier) for this system is given by:

$$
\begin{equation*}
\mathcal{L}(x,\lambda) = x_{1}^{\alpha}x_{2}^{1-\alpha} + \lambda\cdot\left(I-c_{1}x_{1}-c_{2}x_{2}\right)
\end{equation*}
$$

which gives the first-order optimality conditions:

$$
\begin{eqnarray*}
\frac{\partial\mathcal{L}}{\partial{x_{1}}} & = & \alpha\cdot{x_{1}^{\alpha-1}}x_{2}^{1-\alpha} - \lambda\cdot{c_{1}} = 0 \\
\frac{\partial\mathcal{L}}{\partial{x_{2}}} & = & (1-\alpha)\cdot{x_{1}^{\alpha}x_{2}^{-\alpha}} - \lambda\cdot{c_{2}} = 0 \\
\frac{\partial\mathcal{L}}{\partial\lambda} & = & I - c_{1}x_{1} - c_{2}x_{2} = 0
\end{eqnarray*}
$$

This system of first-order optimality conditions can be solved for the optimal consumption levels of $x_{1}$, $x_{2}$, and the [Lagrange multiplier](https://en.wikipedia.org/wiki/Lagrange_multiplier) $\lambda$. However, this is a non-linear system of equations (not easy to find a numerical solution). Thus, we often need to use numerical methods to find the optimal values of $x_{1}$ and $x_{2}$.

## Compute the optimal mix of `Apples` and `Oranges`
To start the computation of the optimal mix of apples versus oranges using a `Cobb-Douglas` utility function, let's first set some values for the problem parameters, namely the $\alpha$ vector, which holds values of the coefficients in the utility function, the $\mathbf{c}$ which holds the unit cost of an apple and an orange, and the `total_budget` that we have to spend:

```julia
# initialize -
α = [0.55, 0.45]; # coefficients
c = [0.98, 1.49]; # price of x1 and x2, we pick these ...
total_budget = 20.0;
```

Next, let's construct an instance of the `MySimpleCobbDouglasChoiceProblem` type using the `build(...)` method. The `MySimpleCobbDouglasChoiceProblem` type, which is encoded in the [src/Types.jl](src/Types.jl) file, holds the problem information. We'll later pass this type to a `solve(...)` function to compute the optimal answer.
* We pass the problem parameters, e.g., the coefficients of the utility function, the cost values, the total budget value, an initial guess, and bounds for the allocation into the `build(...)`. We then pass the `problem` instance to the `solve(...)` method and store the solution in the `solution` variable:

```julia
problem = build(MySimpleCobbDouglasChoiceProblem, (
    
    initial = 0.1*ones(2), # initial guess
    α = α, # coefficients
    c = c, # price of x1 and x2
    I = total_budget, # income
    
    # how much of x₁ and x₂ can be we buy?
    bounds = [
        0.0 100.0; # L U
        0.0 100.0; # L U
    ]
)); # hmmm. why is there an extra set of () inside the function?
```

Finally, pass the `MySimpleCobbDouglasChoiceProblem` problem instance to a `solve(...)` method (implemented in the [src/Solve.jl](src/Solve.jl) file, and return an optimal `solution` dictionary. The `solution` variable is a [Julia Dictionary type](https://docs.julialang.org/en/v1/base/collections/#Base.Dict) which holds three `string keys`:
* `argmax`: Holds the optimal solution vector
* `budget`: Holds the value of resources expended by the optimal solution, e.g., how much money we spent at the optimum
* `objective_value`: Holds the optimal value of the utility

```julia
solution = solve(problem);
```

```julia
solution
```

### Check: Does this solution satisfy the first-order optimality conditions?
If the data in the `solution` is correct, it should satisfy the system of first-order optimality conditions, i.e., it should have a `0` right-hand side. Let's check this using the [@assert macro](https://docs.julialang.org/en/v1/base/base/#Base.@assert) in combination with the [isapprox function](https://docs.julialang.org/en/v1/base/math/#Base.isapprox)
* Let's start this `check` by computing the first-order optimality conditions at the solution estimated by the [MadNLP library](https://github.com/MadNLP/MadNLP.jl), call these `E1,E2` and `E3`

```julia
# get the solution values from the solution dictionary
x̂₁ = solution["argmax"][1];
x̂₂ = solution["argmax"][2];
c₁ = c[1];
c₂ = c[2];
I = solution["budget"]

# compute the marginal utility at the optimum -
Ū₁ = α[1]*(x̂₁^(α[1]-1))*(x̂₂^(1-α[1]))
Ū₂ = (1-α[1])*(x̂₁^(α[1]))*(x̂₂^(-α[1]))

# compute the Lagrange multiplier -
λ = Ū₁/c₁;

# Compute the first-order conditions -
E = zeros(3);
E[1] = Ū₁ - λ*c₁
E[2] = Ū₂ - λ*c₂
E[3] = I - (c₁*x̂₁ + c₂*x̂₂);
```

Next, use a `for` loop to check all the values in the `E` array to some absolute tolerance value, i.e., `atol = 1e-9`.

```julia
for i ∈ eachindex(E)
    @assert isapprox(E[i], 0.0, atol=1e-9)
end
```

## Visualize the Resource Line, Indifference Curve, and the Optimal Solution
`Unhide` the code block below to see how we plotted the solution to the `Cobb-Douglas` maximum utility problem.

```julia
# ok, let's grab the results dictionary, and the Uopt -
Uopt = round(solution["objective_value"], sigdigits=4);
xopt = solution["argmax"];
c = problem.c;
budget_value = problem.I;
xlim = [0.0, 1.1*(budget_value/c[1])];
ylim = [0.0, 1.1*(budget_value/c[2])];
x1_opt = round(xopt[1]; sigdigits=3)
x2_opt = round(xopt[2]; sigdigits=3)

# compute the indifference curve -
INC = indifference(problem, Uopt, xlim)

# compute the budget -
BC = budget(problem, xlim)

# make the plots -
plot(INC[:,1],INC[:,2], lw=3, c=:orange, label="U = $(Uopt) utils",
    bg="floralwhite", background_color_outside="white", framestyle = :box, fg_legend = :transparent);
plot!(BC[:,1],BC[:,2], label="Budget $(budget_value) USD", lw=3,c=:black, xlim = xlim, ylim = ylim)
scatter!([xopt[1]], [xopt[2]], ms=4, c="white", mec=:orange, label="Opt: ($(x1_opt),$(x2_opt))")
xlabel!("Number of Apples", fontsize=18)
ylabel!("Number of Oranges", fontsize=18)
```
